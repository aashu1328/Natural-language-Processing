```
import spacy
sp=spacy.load('en_core_web_sm')
def n_v(text):
  final=sp(text)
  NO_OF_NOUN=0
  NO_OF_VERB=0  
  for word in final:
     if word.pos_=='NOUN':
        NO_OF_NOUN=NO_OF_NOUN+1 
     elif word.pos_=='VERB':
        NO_OF_VERB=NO_OF_VERB+1 
  #return NO_OF_NOUN/NO_OF_VERB
  if(NO_OF_NOUN!=0 and NO_OF_VERB!=0):
    return NO_OF_NOUN/NO_OF_VERB
  else:
    return 0   
```

```
import pandas as pd
import re
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
df = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz', lines=True)    
df.head()
```

```
from nltk.tokenize import MWETokenizer 
from nltk.tokenize import sent_tokenize, word_tokenize
def mwe(str):
    d = {'adjective':0 , 'Noun':0}
    d1 = {}
    tk = MWETokenizer() 
    d['MWETokenizer'] = tk.tokenize(str.split())
    d['word_tokenize'] = word_tokenize(str)
    temp = nltk.pos_tag(str.split())
    d['pos_tag'] = temp

    adj = []
    noun = []
    for x in temp:
        if x[1] == 'JJ':
          d['adjective'] += 1
        elif (x[1] == 'NN' or x[1] == 'NNP' or x[1] == 'NNS' or x[1] == 'NNPS'):
          d['Noun'] +=1
        if d['Noun'] != 0:
          d1 = d['adjective']/d['Noun']
    return d1
```

```
df['ratio']=df['reviewText'].iloc[1:50].apply(mwe)
df.iloc[1:50]
```
